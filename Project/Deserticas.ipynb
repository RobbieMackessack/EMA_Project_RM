```python
import numpy as np
import matplotlib.pyplot as plt

# Define the model parameters
L = 10   # size of the grid (L x L)
p = 0.3   # probability of an initial infection
beta = 0.6   # infection rate
gamma = 0.2   # recovery rate
T = 100   # number of time steps

# Initialize the grid
grid = np.zeros((L, L), dtype=int)
for i in range(L):
    for j in range(L):
        if np.random.random() < p:
            grid[i, j] = 1

# Define the neighborhood function
def neighborhood(i, j):
    """
    Returns the indices of the neighbors of the cell (i, j)
    using a von Neumann neighborhood.
    """
    neighbors = [(i+1, j), (i-1, j), (i, j+1), (i, j-1)]
    neighbors = [(x, y) for x, y in neighbors if 0 <= x < L and 0 <= y < L]
    return neighbors

# Run the simulation
for t in range(T):
    new_grid = np.copy(grid)
    for i in range(L):
        for j in range(L):
            if grid[i, j] == 1:   # if the cell is infected
                # Check if any of the neighbors are susceptible
                for x, y in neighborhood(i, j):
                    if grid[x, y] == 0 and np.random.random() < beta:
                        new_grid[x, y] = 1   # infect the susceptible cell
                # Check if the infected cell recovers
                if np.random.random() < gamma:
                    new_grid[i, j] = 2   # mark the cell as recovered
            elif grid[i, j] == 2:   # if the cell is recovered
                new_grid[i, j] = 0   # reset the cell to susceptible
    grid = np.copy(new_grid)

# Plot the final state of the grid
plt.imshow(grid, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Set up the initial state randomly
def random_state(N):
    return np.random.randint(2, size=(N, N))

# Calculate the next state of the cellular automaton
def update(frameNum, img, state, N):
    # Copy the current state to a new array
    newstate = state.copy()
    # Loop over each cell in the grid
    for i in range(N):
        for j in range(N):
            # Count the number of neighbors that are alive
            # using a Moore neighborhood
            total = int((state[(i-1)%N,(j-1)%N] + state[(i-1)%N,j] + state[(i-1)%N,(j+1)%N] +
                         state[i,(j-1)%N] + state[i,(j+1)%N] +
                         state[(i+1)%N,(j-1)%N] + state[(i+1)%N,j] + state[(i+1)%N,(j+1)%N])/2)
            # Apply the rules of the game
            if state[i,j] == 0:
                if total == 3:
                    newstate[i,j] = 1
            elif total < 2 or total > 3:
                newstate[i,j] = 0
    # Update the state
    img.set_data(newstate)
    state[:] = newstate[:]
    return img,

# Set up the animation
N = 1000
state = random_state(N)
fig, ax = plt.subplots()
img = ax.imshow(state)
ani = animation.FuncAnimation(fig, update, fargs=(img, state, N), frames=10, interval=50, blit=True)
plt.show()
```
